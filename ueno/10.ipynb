```python
import math
from math import gamma
def likelihood(x: int, n: int, psi: float) -> float:
    """
    尤度(二項分布)を計算する関数
    """
    return (math.comb(n, x)) * (psi ** x) * ((1 - psi) ** (n - x))


def posterior_prob(x: int, n: int, alpha: float, beta: float) -> float:
    """
    周辺尤度を計算する関数
    """
    comb = math.comb(n, x)
    deno = gamma(alpha + beta) * gamma(x + alpha) * gamma(n - x + beta)
    nume = gamma(alpha) * gamma(beta) * gamma(n + alpha + beta)
    return comb * (nume / deno)
```
